```python
data = list(map(int, input().split()))

# print(data)

#######################################################
from collections import deque

def dfs(graph, start):
    need, visit = deque(), []
    need.append(start)

    while need:
        node = need.pop()

        if node not in visit:
            visit.append(node)
            need.extendleft(graph[node])  # 역방향으로 큐에 추가

    return visit


def bfs(graph, start):
    visit = []
    need = []

    need.append(start)

    while need:
        node = need.pop(0)
        if node not in visit:
            visit.append(node)
            need.extend(graph[node])
    
    return visit



####################################################

graph = {}

for _ in range(1, data[1]+1):
    # print(1)
    data2 = list(map(int, input().split()))

    if data2[0] not in graph:
        graph[data2[0]] = [data2[1]]
    else:
        graph[data2[0]].append(data2[1])

    if data2[1] not in graph:
        graph[data2[1]] = [data2[0]]
    else:
        graph[data2[1]].append(data2[0])

# print(graph)

print(dfs(graph, data[2]))
print(bfs(graph, data[2]))
```
